# Comparative Clustering of maps across scans

The purpose of this notebook is to run a clustering analysis on the three session per subject. The goal is to cluster all the scans of one subject together. The better this works, the more closely related are the scans of one subject compared to the scans of different subjects. I will also use a scan by scan matrix to visualize the relationship. 

Finally, I will compare the output of matlab and python and see what the differences are.

```python
#Imports
import os
import re
import glob
import numpy as np
import pandas as pd
import nibabel as nib
import brainbox as bb
from matplotlib import pyplot as plt
import scipy.spatial.distance as dist
import scipy.cluster.hierarchy as clh
```

```python
%load_ext pymatbridge
```

## Python define paths and options

```python
#Paths
sub_f = '/data1/scores/retest/out/subjects.txt'
in_path = '/data1/scores/retest/out/sc07/'
in_mask = '/data1/cambridge/template/template_cambridge_basc_multiscale_sym_scale007.nii.gz'
```

```python
# Options
scale = 7
metric = 'dual_regression'
```

## Python grab files

```python
# Get the list of subjects
subs = open(sub_f, 'rb').readline().split(',')
# Grab the files
files = glob.glob(os.path.join(in_path, metric, '*'))
# Store all the files in a dictionary where all the sessions are in a sorted list
sub_dict = dict()
for f in files:
    sub_num = re.search(r'sub[0-9]*', f).group()
    ses_num = int(re.search(r'(?<=session)[0-9]+', f).group())
    if not sub_num in sub_dict.keys():
        sub_dict[sub_num] = ['', '', '']
    sub_dict[sub_num][ses_num - 1] = f

subjects = sub_dict.keys()
num_subs = len(subjects)
num_all = len(files)

# Sort the files into a common order according to the subject file
sort_files = list()
for sub in subs:
    for f in sub_dict[sub]:
        sort_files.append(f)
        
# Make an array of session by voxel in the same order as the subjects and sessions
mask_img = nib.load(in_mask)
mask = mask_img.get_data() > 0
mask_vec = mask.flatten(order='Fortran')
num_vox = np.sum(mask)
# Store all the networks and subjects inside the array
all_arr = np.zeros((num_all, num_vox, scale))

for f_id, f in enumerate(sort_files):
    img = nib.load(f)
    data = img.get_data()
    for net in np.arange(scale):
        net_data = data[..., net]
        net_vec = net_data.flatten(order='Fortran')
        net_mask = net_vec[mask_vec]
        all_arr[f_id, :, net] = net_mask
```

## Matlab define paths and options

```python
%%matlab
% Paths
in_dir = '/data1/scores/retest/out/sc07';
mask_template = '/data1/cambridge/template/template_cambridge_basc_multiscale_sym_scale007.nii.gz';
sub_f = '/data1/scores/retest/out/subjects.txt';
f_sub = fopen(sub_f, 'r');
subjects = textscan(f_sub, repmat('%s', 1, 25), 'delimiter', ',');
num_subs = length(subjects);
[~,~,ext] = niak_fileparts(mask_template);
[m_hdr, m_vol] = niak_read_vol(mask_template);
mask = logical(m_vol);
```

```python
%%matlab
% Options
clusters = 7;
metric = 'stability_maps';
in = [in_dir filesep metric];
```

## Matlab grab files

```python
%%matlab
f = dir(in);
in_strings = {f.name};
in_struct = struct();
f_count = 0;
% Find the files
for f_id = 1:numel(in_strings)
    in_string = in_strings{f_id};
    % Get anything with a nii.gz in the end
    [start, stop] = regexp(in_string, '\w*.nii.gz');
    [sub_start, sub_stop] = regexp(in_string, 'sub[0-9]*');
    [ses_start, ses_stop] = regexp(in_string, 'session[0-9]+');
    if ~isempty(start) && ~isempty(stop)
        f_count = f_count + 1;
        file_name = in_string(start:stop);
        sub_name = in_string(sub_start:sub_stop);
        ses_name = in_string(ses_start:ses_stop);
        in_struct.(sub_name).(ses_name) = [in filesep in_string];
    end
end
% Now load all this and do the cluster and then show it to me
num_subs = length(fieldnames(in_struct));
num_vox = sum(mask(:));
all_array = zeros(num_subs*3, num_vox, clusters);
% Iterate over the subjects
count = 1;
for sub_id = 1:num_subs
    sub = subjects{sub_id}{1};
    sub_struct = in_struct.(sub);
    % Get sessions
    for s_id = 1:3
        s_name = sprintf('session%d',s_id);
        in_file = sub_struct.(s_name);
        [~, vol] = niak_read_vol(in_file);
        for clust = 1:clusters
            net_map = vol(:,:,:,clust);
            net_mask = net_map(mask);
            all_array(count,:,clust) = net_mask;
        end
        count = count + 1;
    end
end
```

```python
# Now for each network, compute the distance metric and linkage and then cluster into 25 clusters
net_part = np.zeros((num_all, num_vox))
net_dist = np.zeros((num_all, num_all, num_vox))
for net in np.arange(scale):
    distance = dist.squareform(dist.pdist(all_arr[..., net], 'euclidean'))
    linkage = clh.linkage(distance, method='ward')
    part = clh.fcluster(linkage, 25, criterion='maxclust')
    net_part[:, net] = part
    net_dist[..., net] = distance
```

```python
# Visualize the distance matrices
f = plt.figure(figsize=(15,30))
for net in np.arange(scale):
    ax = f.add_subplot(4,2,net+1)
    ax.set_xticks([])
    ax.set_yticks([])
    subdist = bb.visu.add_subplot_axes(ax, [0, 0, 0.9, 1])
    subcol = bb.visu.add_subplot_axes(ax, [0.91, 0, 0.09, 1])
    im = subdist.matshow(net_dist[..., net], cmap=cm.Blues)
    subdist.set_xticks([])
    subdist.set_yticks([])
    plt.colorbar(im, cax=subcol)
    ax.set_title('Network {}'.format(net+1))
```

```python
# Now show the clustering for each network
f = plt.figure(figsize=(15,30))
for net in np.arange(scale):
    ax = f.add_subplot(4,2,net+1)
    ax.set_xticks([])
    ax.set_yticks([])
    part = net_part[:, net]
    
    tmp = np.zeros((75,75))
    # Turn the clustering into a matrix
    for cl_id in np.arange(25):
        cl_val = cl_id + 1
        tmp[np.ix_(np.where(part==cl_val)[0], np.where(part==cl_val)[0])] = cl_id
        
    subdist = bb.visu.add_subplot_axes(ax, [0, 0, 0.9, 1])
    subcol = bb.visu.add_subplot_axes(ax, [0.91, 0, 0.09, 1])
    im = subdist.matshow(tmp)
    subdist.set_xticks([])
    subdist.set_yticks([])
    plt.colorbar(im, cax=subcol)
    ax.set_title('Network {}'.format(net+1))
```
