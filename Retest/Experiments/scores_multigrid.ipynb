# The purpose of this notebook is to make quick and easy montages of the images generated by matlab

I want to have all three metrics in rows and the three sessions in columns. This way, I will have only one figure per network and scale

```python
import os
import re
import glob
import Image
from matplotlib import pyplot as plt
```

```python
# Parameters
scales = [7]
sessions = ['12', '13', '23']
metrics = ('stability_maps', 'rmap_part', 'dual_regression')
in_path = '/data1/scores/retest/out/newtest/'
in_temp = os.path.join(in_path, 'sc{:02d}', '{}', 'figures', '{}');
fig_dir = '/data1/scores/retest/out/newtest/figures'
if not os.path.isdir(fig_dir):
    os.makedirs(fig_dir)
```

```python
# Get all the figures in the current folder and sort them by network and session
scale_list = list()
for scale in scales:
    file_dict = dict()
    for m_id, metric in enumerate(metrics):
        images = glob.glob(in_temp.format(scale, metric, '*.png'))
        for image in images:
            network = int(re.search('(?<=icc_)[0-9]+',image).group())
            session = re.search('[0-9]+(?=_{})'.format(metric),image).group()
            if not network in file_dict.keys():
                file_dict[network] = [[None]*3, [None]*3, [None]*3]
            for s_id, ses in enumerate(sessions):
                if session == ses:
                    file_dict[network][m_id][s_id] = image
                else:
                    continue
    scale_list.append(file_dict)
```

```python
# Make the figures
for sc_id, scale in enumerate(scales):
    for n_id in np.arange(scale):
        network = n_id + 1
        paths = scale_list[sc_id][network]
        f = plt.figure(figsize=(40,30))
        f_name = os.path.join(fig_dir, 'net_{}_scale_{}.png'.format(scale, network))
        place = np.reshape(np.arange(1,10), (3,3))
        for m_id, metric in enumerate(metrics):
            for s_id, ses in enumerate(sessions):
                image = paths[m_id][s_id]
                pos = place[m_id, s_id]
                ax = f.add_subplot(3,3,pos)
                ax.set_axis_off()
                i = Image.open(image)
                ax.imshow(i)
                plt.tight_layout()
        f.savefig(f_name, bbox_inches='tight')
        f.clear()
```
