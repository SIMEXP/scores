# 02/03/2015 - Bad Priors

The purpose of this experiment is:
* to investigate the effect of bad priors on the stability estimates of scores
* to compare the performance of seed based connectivity to the same data
* explore this across a range of noise and structured noise parameters

### Get the environment started up

```python
%load_ext pymatbridge
```

## Generate a simulated timeseries

I will generate a simulated timeseries with properties that I am usually using to explore seed metrics. 
First, I will set these parameters here:

```python
%%matlab
edge = 32;
target_net = 2;

opt_s.type = 'checkerboard'; 
opt_s.t = 100; 
opt_s.n = edge*edge; 
opt_s.nb_clusters = [4 16]; 
opt_s.fwhm = 1; 
opt_s.variance = 0.05;
```

The target_net parameter here defines the net in the prior that I want to use both for the seed based correlation and from which I will show the scores based stability map.

Now to simulate a timeseries based on the specified checkerboard, I run [niak_simus_scenario](https://github.com/SIMEXP/niak/blob/master/commands/simus/niak_simus_scenario.m)

```python
%%matlab
[tseries,opt_mplm] = niak_simus_scenario(opt_s);
```

Let's have a quick look at the correlation matrix of this time series

```python
%%matlab
R = corr(tseries);
hier = niak_hierarchical_clustering(R);
order = niak_hier2order(hier);
niak_visu_matrix(R(order, order));
```

Let's get the true partition that was used to generate the time series (the 4 by 4 version with 16 elements) and have a look at that.

```python
%%matlab
prior = opt_mplm.space.mpart{2};
prior_2d = reshape(prior, [edge, edge]);

figure;
niak_visu_matrix(prior_2d);
```

Looks good. The networks are labeled from the to left corner so the target network can be identified by the colors.

Now, let's roll these priors around so they are somewhat distorted and then get them back into vector form, so we can use them. But also have a look at them

```python
%%matlab
shifted_prior_2d = circshift(prior_2d, 3, 1);
shifted_prior_2d = circshift(shifted_prior_2d, 3, 2);
shifted_prior = reshape(shifted_prior_2d, [dot(edge, edge), 1]);

figure;
niak_visu_matrix(shifted_prior_2d);
```

Alright, so things have moved around a bit. Let's see how scores deals with the noise under the good and the shifted prior.  
I will run [niak_stability_cores](https://github.com/SIMEXP/niak/blob/scores/extensions/surfstab/niak_stability_cores.m) with the time series simulated above and hand over the two different priors.

(The double Matlab/Python comments are just for the notebook)

```python
%%matlab
opt_scores.sampling.type = 'bootstrap';
opt_scores.sampling.opt = opt_s;
%# Scores with correct prior
res_good = niak_stability_cores(tseries,prior,opt_scores);
scores_good = reshape(res_good.stab_maps(:, target_net), [32 32]);
%# Scores with bad prior
res_shift = niak_stability_cores(tseries,shifted_prior,opt_scores);
scores_shift = reshape(res_shift.stab_maps(:, target_net), [32 32]);
```

Now let's have a look at the output

```python
%%matlab
subplot(1,2,1);
niak_visu_matrix(scores_good);
title('scores with good prior');
subplot(1,2,2);
niak_visu_matrix(scores_shift);
title('scores with shifted prior');
```

Ok, so under this pretty noisy data, scores does get affected by the shifted prior. What about the __seed based__ method?

```python
%%matlab
target_good = prior == target_net;

%# With the good prior
seed_sig_good = mean(tseries(:,target_good),2);
seed_vec_good = corr(seed_sig_good, tseries);
seed_map_good = reshape(seed_vec_good, [32 32]);

target_shift = shifted_prior == target_net;

%# With the shifted prior
seed_sig_shift = mean(tseries(:,target_shift),2);
seed_vec_shift = corr(seed_sig_shift, tseries);
seed_map_shift = reshape(seed_vec_shift, [32 32]);
```

Let's take a look

```python
%%matlab
subplot(2,2,1);
niak_visu_matrix(scores_good);
title('scores with good prior');
subplot(2,2,2);
niak_visu_matrix(scores_shift);
title('scores with shifted prior');

subplot(2,2,3);
niak_visu_matrix(seed_map_good);
title('seed with good prior');
subplot(2,2,4);
niak_visu_matrix(seed_map_shift);
title('seed with shifted prior');
```

Seed is not doing a lot better. In fact, it looks a lot worse. But it's not so easy to tell with the shitty 'one color scale fits all' behaviour of matlab figures.  

### With better signal.

```python
%%matlab
opt_s.fwhm = 1; 
opt_s.variance = 0.5;

[tseries_good,~] = niak_simus_scenario(opt_s);
```

```python
%%matlab
%# First the scores technique
opt_scores.sampling.type = 'bootstrap';
opt_scores.sampling.opt = opt_s;
%# Scores with correct prior
res_g_g = niak_stability_cores(tseries_good,prior,opt_scores);
scores_g_g = reshape(res_g_g.stab_maps(:, target_net), [32 32]);
%# Scores with bad prior
res_s_g = niak_stability_cores(tseries_good,shifted_prior,opt_scores);
scores_s_g = reshape(res_s_g.stab_maps(:, target_net), [32 32]);

%# And then with the seed based method
target_good = prior == target_net;
%# With the good prior
seed_sig_g_g = mean(tseries_good(:,target_good),2);
seed_vec_g_g = corr(seed_sig_g_g, tseries_good);
seed_map_g_g = reshape(seed_vec_g_g, [32 32]);


target_shift = shifted_prior == target_net;
%# With the shifted prior
seed_sig_s_g = mean(tseries_good(:,target_shift),2);
seed_vec_s_g = corr(seed_sig_s_g, tseries_good);
seed_map_s_g = reshape(seed_vec_s_g, [32 32]);
```

How does __scores__ fare against __seed__ under __better signal__

```python
%%matlab
subplot(2,2,1);
niak_visu_matrix(scores_g_g);
title('scores with good prior');
subplot(2,2,2);
niak_visu_matrix(scores_s_g);
title('scores with shifted prior');

subplot(2,2,3);
niak_visu_matrix(seed_map_g_g);
title('seed with good prior');
subplot(2,2,4);
niak_visu_matrix(seed_map_s_g);
title('seed with shifted prior');
```

### Look at the silhouette maps for scores

```python
%%matlab
scores_sil_g = reshape(res_g_g.stab_contrast, [32 32]);
scores_sil_s = reshape(res_s_g.stab_contrast, [32 32]);
```

```python
%%matlab
subplot(2,2,1);
niak_visu_matrix(scores_g_g);
title('stab with good prior');
subplot(2,2,2);
niak_visu_matrix(scores_s_g);
title('stab with shifted prior');

subplot(2,2,3);
niak_visu_matrix(scores_sil_g);
title('sil with good prior');
subplot(2,2,4);
niak_visu_matrix(scores_sil_s);
title('sil with shifted prior');
```

Interesting.  
This looks exactly as if the border subclusters have their stability stolen by the next cluster. Everything on the border has really bad silhouette values. Let's take a quick look at a network that sits in the corner - like network 1 for example.

```python
%%matlab
comp_s = reshape(res_s_g.stab_maps(:, 1), [32 32]);
comp_g = reshape(res_g_g.stab_maps(:, 1), [32 32]);

subplot(2,2,1);
niak_visu_matrix(scores_g_g);
title('Good border prior');
subplot(2,2,2);
niak_visu_matrix(scores_s_g);
title('Shifted border prior');

subplot(2,2,3);
niak_visu_matrix(comp_g);
title('Good corner prior');
subplot(2,2,4);
niak_visu_matrix(comp_s);
title('Shifted corner prior');
```

This is pretty amazing. It is pretty clear from these plots, that scores does not bleed into the next cluster with which it does not have any relationship - expcept for the noise. However, if we take the corner prior and shift it, it will overlap with other subclusters from the same cluster and bleed into them - because it has a lot of shared signal.  
In the case of the border prior, this is a different story. The portion of them, that overlaps with the next cluster will usually be assigned to the prior of the other cluster because it still shares more signal with it (i.e. the overlap of the prior with the next cluster is less than that of the next prior). So this part will be stolen away and have low stability. And on the left, the corner prior is also stealing stability from the prior, because they share a lot of signal and because the shifted prior is also picking up the signal from the next cluster.

Now I wonder what would happen, if I shifted the priors one half of their own width. They would then overlap 50% with the next network and the effect should get even stronger. But for now, I will just resimulate the timeseries with more time points to get a better idea.

### Re-generate the time series with more time points to get a clearer idea of what is happening at the borders

```python
%%matlab
opt_s.t = 3000;
opt_s.fwhm = 1; 
opt_s.variance = 0.5;

[tseries_long,~] = niak_simus_scenario(opt_s);
```

```python
%%matlab
%# First the scores technique
opt_scores.sampling.type = 'bootstrap';
opt_scores.sampling.opt = opt_s;
%# Scores with correct prior
res_g_l = niak_stability_cores(tseries_long,prior,opt_scores);
scores_g_l = reshape(res_g_l.stab_maps(:, target_net), [32 32]);
%# Scores with bad prior
res_s_l = niak_stability_cores(tseries_long,shifted_prior,opt_scores);
scores_s_l = reshape(res_s_l.stab_maps(:, target_net), [32 32]);
```

```python
%%matlab
comp_s_l = reshape(res_s_l.stab_maps(:, 1), [32 32]);
comp_g_l = reshape(res_g_l.stab_maps(:, 1), [32 32]);

scores_sil_g_l = reshape(res_g_l.stab_contrast, [32 32]);
scores_sil_s_l = reshape(res_s_l.stab_contrast, [32 32]);

subplot(2,2,1);
niak_visu_matrix(scores_g_l);
title('stab with good prior');
subplot(2,2,2);
niak_visu_matrix(scores_s_l);
title('stab with shifted prior');

subplot(2,2,3);
niak_visu_matrix(scores_sil_g_l);
title('sil with good prior');
subplot(2,2,4);
niak_visu_matrix(scores_sil_s_l);
title('sil with shifted prior');
```

Ok, this is pretty obvious now. I think the shifted corner priors are stealing stability of the border priors but the border priors can only steal from the other border priors in the same cluster. Because if the prior overlaps with a different cluster, then the prior of the different cluster is going to keep all the voxels inside that (true) cluster because they share no signal with the shifted prior aside from the bit it picks up by being shifted. And because the prior of the new cluster still has a lot of overlap with it's own cluster (even if it would be pushed one half of it's width, there is still the shared signal) it will usually win the voxels over.

This is pretty much very good.

Now I don't think we even have to see what seed does here but I will still take a look. The more relevant question is however: __what happens if the size of the prior does no longer match the structure in the data__, for example because it encompasses the entire cluster. Then shifting should be more messy.

So first: seed with the same long time series:

```python
%%matlab

%# And then with the seed based method
target_good = prior == 2;
%# With the good prior
seed_sig_g_l = mean(tseries_long(:,target_good),2);
seed_vec_g_l = corr(seed_sig_g_l, tseries_long);
seed_map_g_l_border = reshape(seed_vec_g_l, [32 32]);


target_shift = shifted_prior == 2;
%# With the shifted prior
seed_sig_s_l = mean(tseries_long(:,target_shift),2);
seed_vec_s_l = corr(seed_sig_s_l, tseries_long);
seed_map_s_l_border = reshape(seed_vec_s_l, [32 32]);

%# And then with the seed based method
target_good = prior == 1;
%# With the good prior
seed_sig_g_l = mean(tseries_long(:,target_good),2);
seed_vec_g_l = corr(seed_sig_g_l, tseries_long);
seed_map_g_l_corner = reshape(seed_vec_g_l, [32 32]);


target_shift = shifted_prior == 1;
%# With the shifted prior
seed_sig_s_l = mean(tseries_long(:,target_shift),2);
seed_vec_s_l = corr(seed_sig_s_l, tseries_long);
seed_map_s_l_corner = reshape(seed_vec_s_l, [32 32]);
```

```python
%%matlab
subplot(2,2,1);
niak_visu_matrix(seed_map_g_l_corner);
title('corner seed with good prior');
subplot(2,2,2);
niak_visu_matrix(seed_map_s_l_border);
title('border seed with good prior');

subplot(2,2,3);
niak_visu_matrix(seed_map_s_l_corner);
title('corner seed with shifted prior');

subplot(2,2,4);
niak_visu_matrix(seed_map_s_l_border);
title('border seed with shifted prior');
```

Alright, so the average outside of the correct cluster is pretty much zero. However, with the shifted prior this is a different story. Seed bleeds happily into the next cluster. So my guess is: if I choose the prior as something that is at the border of all clusters, we should see a very big map

```python
%%matlab

%# And then with the seed based method
target_good = prior == 6;
%# With the good prior
seed_sig_g_l = mean(tseries_long(:,target_good),2);
seed_vec_g_l = corr(seed_sig_g_l, tseries_long);
seed_map_g_l_border = reshape(seed_vec_g_l, [32 32]);


target_shift = shifted_prior == 6;
%# With the shifted prior
seed_sig_s_l = mean(tseries_long(:,target_shift),2);
seed_vec_s_l = corr(seed_sig_s_l, tseries_long);
seed_map_s_l_border = reshape(seed_vec_s_l, [32 32]);

opt.limits = [-1 1];
opt.color_map = 'hot_cold';

subplot(2,2,1);
niak_visu_matrix(seed_map_g_l_corner, opt);
title('corner seed with good prior');
subplot(2,2,2);
niak_visu_matrix(seed_map_g_l_border, opt);
title('border seed with good prior');

subplot(2,2,3);
niak_visu_matrix(seed_map_s_l_corner, opt);
title('corner seed with shifted prior');

subplot(2,2,4);
niak_visu_matrix(seed_map_s_l_border, opt);
title('border seed with shifted prior');
```

Yeah, pretty much that. So not a good idea to have a mismatched prior with seed - this will give you a lot of annoying results. Just as a comparison, I'll put the corresponding scores next to it.

```python
%%matlab
corner_scores_s = reshape(res_s_g.stab_maps(:, 1), [32 32]);
corner_scores_g = reshape(res_g_g.stab_maps(:, 1), [32 32]);

border_scores_s = reshape(res_s_g.stab_maps(:, 6), [32 32]);
border_scores_g = reshape(res_g_g.stab_maps(:, 6), [32 32]);

%#Need two figures here because matlab is retarded
figure;
subplot(2,2,1);
niak_visu_matrix(abs(seed_map_g_l_corner));
title('corner seed with good prior');
subplot(2,2,2);
niak_visu_matrix(abs(seed_map_g_l_border));
title('border seed with good prior');
subplot(2,2,3);
niak_visu_matrix(corner_scores_g);
title('corner scores with good prior');
subplot(2,2,4);
niak_visu_matrix(border_scores_g);
title('border scores with good prior');

figure;
subplot(2,2,1);
niak_visu_matrix(abs(seed_map_s_l_corner));
title('corner seed with shifted prior');
subplot(2,2,2);
niak_visu_matrix(abs(seed_map_s_l_border));
title('border seed with shifted prior');
subplot(2,2,3);
niak_visu_matrix(corner_scores_s);
title('corner cores with shifted prior');
subplot(2,2,4);
niak_visu_matrix(border_scores_s);
title('border scores with shifted prior');
```

Alright, this is a nice overview of what happens with misaligned priors if the priors match the true signal in size. Bottom line is: __scores blows seed out of the water__
