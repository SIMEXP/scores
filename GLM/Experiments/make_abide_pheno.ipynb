```python
import re
import os
import glob
import pandas as pd
```

```python
motion_pheno = '/data1/abide/Pheno/motion_all.csv'
other_pheno = '/data1/abide/Pheno/pheno.csv'
out_pheno = '/data1/abide/Pheno/glm_pheno.csv'
```

```python
mot = pd.read_csv(motion_pheno)
phe = pd.read_csv(other_pheno)
```

```python
# Make a combined file

with open(out_pheno, 'w') as outfile:
    outfile.write('sub,site,sex,diag,age,mean_fd\n')
    for index, row in mot.iterrows():
        sub_name = int(re.search('(?<=X00)[0-9]*', row[' ']).group())
        ses_name = re.search('_session_1_run1', row[' '])
        if not ses_name:
            print('nah', row[' '])
            continue
        FD = row['FD_scrubbed']
        phe_row = phe.loc[phe['SUB_ID'] == sub_name]
        sex = phe_row['SEX'].values[0]
        site = phe_row['SITE_ID'].values[0]
        diag = phe_row['DX_GROUP'].values[0]
        age = phe_row['AGE_AT_SCAN'].values[0]
        line = '{},{},{},{},{},{}\n'.format(sub_name, site, sex, diag, age, FD)
        outfile.write(line)
```
